```python
# Step 1: Importing necessary libraries
import pandas as pd  # For handling data
import numpy as np  # For numerical operations
from sklearn.linear_model import LinearRegression  # For Linear Regression model
from sklearn.metrics import mean_squared_error, r2_score  # For evaluating the model
```

```python
# Step 2: Loading the dataset
data = pd.read_csv('advertising.csv')
```

```python
# Step 3: Data Exploration
print("First few rows of the dataset:")
print(data.head())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel('TV Advertisement')
plt.ylabel('Sales')
plt.scatter(data.TV,data.Sales, color='red', marker= '+')
```

```python
# Basic summary statistics of the dataset
print("\nSummary statistics:")
print(data.describe())

# Set 1: [50, 60, 70, 80, 90]
# Set 2: [55, 65, 75, 85, 95]

# Both sets have the same average score of 70. However, the scores in Set 1 are more spread out (they range from 50 to 90), 
# while the scores in Set 2 are closer to 70 (they range from 55 to 95).
```

```python
# Step 4: Feature Selection and Data Preparation
X = data[['TV', 'Radio', 'Newspaper']]  # Independent variables (features)
y = data['Sales']  # Dependent variable (target)
```

```python
# Step 5: Model Building and Training
# Initializing the Linear Regression model
model = LinearRegression()
```

```python
# Training the model using the entire dataset (No train-test split)
model.fit(X, y)
```

```python
# Step 6: Making Predictions
# Predicting the sales based on the advertising budgets
y_pred = model.predict(X)
```

```python
# Step 7: Model Evaluation
# Calculating the Mean Squared Error (MSE) between actual and predicted values
mse = mean_squared_error(y, y_pred)
# Calculating the R-squared (R²) value to check model fit
r2 = r2_score(y, y_pred)
```

```python
# Printing model evaluation metrics
print("\nModel Evaluation Metrics:")
print("Mean Squared Error:", mse) # The MSE value of 2.706 means that, on average, the squared difference between the predicted values and the actual values is 2.706.
print("R-squared:", r2) # 0.9026 suggests that the model has a strong relationship with the target variable and is quite effective at predicting sales.
```

```python
# Step 8: Model Coefficients and Intercept
# Displaying the coefficients and intercept of the linear regression model
print("\nModel Coefficients:")
print("TV Coefficient:", model.coef_[0])
print("Radio Coefficient:", model.coef_[1])
print("Newspaper Coefficient:", model.coef_[2])
print("Intercept:", model.intercept_)
```
